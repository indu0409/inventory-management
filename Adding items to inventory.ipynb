# Adding new  items to the inventory

```python
import json
fd=open("record.json","r")
r=fd.read()
fd.close()
di=json.loads(r)
```

```python
di
```

```python
k=di.keys()
k
```

## adding quantites to an existing item in the inventory

```python
p_id=input("enter the product id:\n")
p_quantity=int(input("enter the quantites of the product:\n"))
if p_id in k:
    di[p_id]['quantity']=di[p_id]['quantity']+p_quantity
else:
    p_name=input("enter the name of the product:\n")
    p_category=input("enter the category of the product:\n")
    p_price=int(input("enter the price of the product:\n"))
    p_brand=input("enter the brand of the product:\n")
    di[p_id]={"name":p_name,"category":p_category,"price":p_price,"quantity":p_quantity,"brand":p_brand}
js=json.dumps(di)
fd=open("record.json","w")
fd.write(js)
fd.close()
```
